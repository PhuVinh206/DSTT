```python
import numpy as np
A = np.array([0,0,-1,2,3,-1,0,1,-1,-1])
search_vector = np.array([0,1,-1])
len(A), len(search_vector) # nghĩa là = (10, 3)
```

```python
B = np.array([1])
B = np.resize(B, (len(A)-len(search_vector)+1, len(search_vector)))
B = np.asmatrix(B)
for i in range(len(A)-len(search_vector)+1): # so dong
 for j in range(len(search_vector)): # so cot
  B[i,j] = A[i+j]
print (B)
print (A)
```

```python
from PIL import Image, ImageDraw
from google.colab import files
uploaded = files.upload()
input_image = Image.open("traidat (3).png")
input_pixels = input_image.load()
box_kernel = [[1/9, 1/9, 1/9],
              [1/9, 1/9, 1/9],
              [1/9, 1/9, 1/9]]
kernel = box_kernel
offset = len(kernel) // 2
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * kernel[a][b]
                acc[1] += pixel[1] * kernel[a][b]
                acc[2] += pixel[2] * kernel[a][b]
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))
output_image.save("blurred.png")
from IPython.display import display
display(output_image)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
gaussian_kernel = (1/256) * np.array([
    [1, 4, 6, 4, 1],
    [4,16,24,16,4],
    [6,24,36,24,6],
    [4,16,24,16,4],
    [1, 4, 6, 4, 1]
], dtype=np.float32)
img = cv2.imread('meo bi bat.jpg')
if img is None:
    print("Không load được ảnh.")
else:
    blurred = cv2.filter2D(img, -1, gaussian_kernel)
    cv2_imshow(blurred)
```

```python
from PIL import Image, ImageDraw
from google.colab import files
uploaded = files.upload()
from IPython.display import display
input_image = Image.open("uia.jpg").convert("L")
input_pixels = input_image.load()
kernels = {
    'kernel_a': [[0, -1,  0],
                 [-1, 4, -1],
                 [0, -1,  0]],
    'kernel_b': [[-1, -1, -1],
                 [-1, 8, -1],
                 [-1, -1, -1]]
}
offset = 1
for name, kernel in kernels.items():
    output_image = Image.new("L", input_image.size)
    draw = ImageDraw.Draw(output_image)
    for x in range(offset, input_image.width - offset):
        for y in range(offset, input_image.height - offset):
            acc = 0
            for a in range(3):
                for b in range(3):
                    xn = x + a - offset
                    yn = y + b - offset
                    pixel = input_pixels[xn, yn]
                    acc += pixel * kernel[a][b]
            acc = max(0, min(255, int(acc)))
            draw.point((x, y), acc)
    output_image.save(f"{name}_edge.png")
    print(f"Ảnh lọc theo {name}:")
    display(output_image)
```
